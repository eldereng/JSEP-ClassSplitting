```python
import pandas as pd

Columns=['SYSTEM', 'RELEASE', 'CLASS', 'WMC', 'DIT', 'NOC', 'CBO', 'RFC', 'CA', 'CE', 'LCOM',
         'NPM', 'LCOM3', 'LOC', 'DAM', 'MOA', 'MFA', 'CAM', 'IC', 'CBM', 'AMC', ]

# Dataset from "ExtractPositiveClasses.R"
df_PositiveCls = pd.read_csv('PositiveClasses.csv', delimiter=",", decimal=".")
df_PositiveCls["REFACT"]=1.0
df_PositiveCls=df_PositiveCls[Columns+["REFACT"]]

# Random Classes extracted from "MetricsPerClassReleaseOrdered.csv" and classified as "Negative" by rules of R Script.
df_NegativeCls = pd.read_csv('NegativeClasses.csv', delimiter=",", decimal=".")
df_NegativeCls["REFACT"]=0.0
df_NegativeCls=df_NegativeCls[Columns+["REFACT"]]

# Dataset used to Create Decision Tree Classifier
FinalDataSet=pd.concat([df_NegativeCls, df_PositiveCls], ignore_index=True)
FinalDataSet.to_csv("DatasetBalancedRefactFinal.csv", sep=",", decimal=".",  index=False)

print("Data Size: ",FinalDataSet.shape, 
      "Class 1.0 (Refact): ", len(FinalDataSet[FinalDataSet.REFACT==1.0]), 
      "Class 0.0 (not Refact): ", len(FinalDataSet[FinalDataSet.REFACT!=1.0]), )
FinalDataSet.head()
```

```python
## Subject Systems: releases and classes - Table 1
df_ALL = pd.read_csv("MetricsPerClassReleaseOrdered.csv", delimiter=";", decimal=".")
RELEASE=df_ALL[["SYSTEM","RELEASE"]].groupby(["SYSTEM"]).agg( {"RELEASE": lambda x: x.nunique()} )
CLASS=df_ALL[["SYSTEM","CLASS"]].groupby(["SYSTEM"]).agg( {"CLASS": lambda x: x.nunique()} )
pd.concat([RELEASE, CLASS], axis=1)
```
